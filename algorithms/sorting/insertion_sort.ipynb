# Insertion Sort

Given an array of length n of unsorted values:

1. Start "left" at index 1 as if this is your first seen value of the array.
2. Then, step-by-step move "right" through the array:
3. Compare each "new" value that you thus come across with the ones you have already seen going right to left, from high to low value.
4. INSERT the "new" value at the correct place, if necessary shifting values to the right to make space.
5. Repeat untill all values are in their right place.
   

```python
A = [5, 2, 4, 6, 1, 3]

def insertion_sort(A):
    n = len(A)
    
    # We traverse the array
    # take the value at the address (index)
    # compare the value with the previous
    # to find the address that will locally sort the array
    for j in range(1, n):
        
        key = A[j]
        
        print(A,'\n',' '*(3*j)+'k')
        
        i = j - 1
    
        # Local comparison
        # While there are values to compare with (i >= 0)
        # and we see a value higher than our key
        while(i >= 0):
    
            print(A,'\n',' '*(3*i)+'^')    
            
            if A[i] > key:
    
                A[i+1] = A[i] # move a value right
    
                print(A, '\n',''*(1+3*i)+'-'+' /')
                
                i -= 1 # inspect value left
            else:
                break
                
        A[i+1] = key
        
        print(A,'\n',' '*(3*(i+1))+'i')
        
        print('\n')

insertion_sort(A)
```
